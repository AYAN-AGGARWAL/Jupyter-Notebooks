### What is Deepface?
Deepface is a lightweight face recognition and facial attribute analysis (age, gender, emotion and race) framework for python.  
It is a hybrid face recognition framework wrapping state-of-the-art models such as:
- VGG-Face 
- Google FaceNet 
- OpenFace 
- Facebook DeepFace
- DeepID
- ArcFace
- Dlib 
- SFace.
> These models are pre-trained 

Deepface provides various features such as:
- Face recognition
- Verifying similarities
- Age Prediction
- Gender Prediction
- Race Prediction
- Emotion Prediction   
and more...

#### Function to display the image (using matplotlib) and predict the age (using deepface):

```python
def predict(name):
    location = r"C:\Users\super\OneDrive\Documents\20\\" + name
    # location = r"C:\Users\super\Software\Mega\Academics\5th sem labs\DS Lab\Datasets\Train\\" + name
    result = DeepFace.analyze(location, actions=['age'], enforce_detection=False) 
    img = mpimg.imread(location)
    plt.figure(figsize=(1, 2))
    imgplot = plt.imshow(img)
    plt.show()
    print(f"Predicted Age: {result['age']}")
    print("___________________________________________")
```

```python
#to check for number of images printed:
count = 0
for i in range(0,1000):
    print(f"Image number {i+1}")
    predict(arr[i])
    count += 1
    #stop printing after 10 images:
    if count == 10:
        break
```
